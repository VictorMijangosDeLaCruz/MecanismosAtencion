# Visual Transformers

Los transformadores visuales (o ViT de Visual Transformer) son un arquitectura de tipo encoder-only de un transformador que se orientan al procesamiento de imágenes, principalmente para tareas de clasificación. La idea central de los ViT es tratar a las imágenes como si fueran textos. Para esto generan <i>patches</i> de la imagen, que son secciones de la imagen que funcionan como los tókens en el texto. Aquí revisamos una implementación básica de los ViT con Pytorch.

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from transformers import *
from copy import deepcopy
from tqdm import tqdm
```

## Implementación de ViT

En primer lugar, necesitamos los mecanismos de self-attention, además como la codificación, tanto los embeddings, como la codificación posicional. Para la atención, utilizamos la implementación típica dada como:
$$Att(Q,K,V) = Softmax\Big( \frac{QK^T}{\sqrt{d}} V\Big)$$
Esta implementación es la ya utilizada previamente en esta <a href="https://victormijangosdelacruz.github.io/MecanismosAtencion/html/02SelfAttention.html">liga</a>. Asimismo, definimos la <a href="https://victormijangosdelacruz.github.io/MecanismosAtencion/html/04Encoding.html">codificación posicional</a> como ya es conocido.

```python
class SelfAttention(nn.Module):
    def __init__(self, d_model):
        super(SelfAttention, self).__init__()
        # Capas de proyecciones
        self.d_model = d_model
        self.Q = nn.Linear(d_model, d_model, bias=False)
        self.K  = nn.Linear(d_model, d_model, bias=False)
        self.V  = nn.Linear(d_model, d_model, bias=False)
        
    def forward(self, x):
        # Proyección de los datos
        query,key,value = self.Q(x),self.K(x),self.V(x)
        # Cálculo de pesos de atención
        scores = torch.matmul(query, key.transpose(-1,-2))/np.sqrt(self.d_model)
        p_attn = torch.nn.functional.softmax(scores, dim = -1)
        #Suma ponderada
        Vs = torch.matmul(p_attn, value).reshape(x.shape)
        
        return Vs, p_attn
    
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=1000, warmup=10000):
        super(PositionalEncoding, self).__init__()
        self.d_model = d_model
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2) * -(torch.log(torch.tensor(warmup)) / d_model))

        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
                         
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        batch_size, seq_len, _ = x.shape
        pe_slice = self.pe[:seq_len, :self.d_model]

        return pe_slice.unsqueeze(0).expand(batch_size, -1, -1)
```

### Arquitectura ViT

El punto central de ViT está en la arquitectura que se define. En primer lugar, se tiene que definir los patches.
Los <b>patches</b> funcionan como los tokens en el procesamiento de texto. Se trata de regiones de la imagen que se codifican para representar la imagen a partir de la secuencia de estas regiones.

Dada una imagen $x \in \mathbb{R}^{H\times W \times C}$, donde $H$ es altura, $W$ anchura y $C$ número de canales, se genera un número dado de patches $x_p \in \mathbb{R}^{N\times P^2C}$ donde $P$ es el tamaño de los patches y $N$ es el total de patches obtenidos de la imagen que puede calcularse como $N = \frac{HW}{P}$ ($P$ debe dividir a $HW$). La obtención de estos patches se puede realizar a partir de una ventana de tamaño $P\times P$ sobre la imagen, o bien como lo hacemos aquí, a partir del uso de una capa convolucional con un kernel de tamaño $P$ que se mueve $P$ pasos (stride). El uso de esta capa convolucional simula la aplicación de un embedding, pues se observa que cada patch $x_p$ debe aplicársele un embedding:
$$e_p = W^e x_p$$
De tal forma, que se tendrá una secuencia de vectores $e_p$ a los que además se les sumará una codificación posicional.
$$z_p = e_p + pe_i$$
Donde $pe_i$ es la codificación de la posición $i$. Finalmente, se creará un vector de clasificación <tt>CLS</tt> (de forma similar a BERT) que se agregará a los vectores codificados. Este vector será el que se use para la clasificación. Entonces, la representación de los vectores será:
$$\mathbf{z} = \{cls, z_1,...,z_n\}$$
donde $cls$ es el vector de clasificación. Posteriormente, se pasará a una capa de auto-atención:
$$\mathbf{h}= Self-Att(\mathbf{z})$$
Y finalmente pasarán a una feedforward (usualmente de profundidad 2 con activación GeLU). En cada paso se agrega <a href="https://victormijangosdelacruz.github.io/MecanismosAtencion/html/05Normalization.html">suma y normalización</a>.  Visualmente, la arquitectura de la red ViT se puede resumir como:

<img src="images/ViT.png" width="450" class="center">

El elemento 0 es el vector de clasificación y los otros elementos son los patches que representan regiones de la imagen.

```python
 class ViT(nn.Module):
    def __init__(self, image_size, patch_size, num_classes, dim, heads, mlp_dim=256, channels=1, p=0.3):
        super().__init__()
        
        assert image_size % patch_size == 0, 'La dimensión de las imágenes debe ser divisible entre tamaño de pathc'
        
        num_patches = (image_size // patch_size)**2
        patch_dim = channels*patch_size**2
        
        self.dim = dim
        #Embedding de pactehs, secciones de imágenes
        self.to_patch_embedding = nn.Sequential(nn.Conv2d(channels, dim, kernel_size=patch_size, stride=patch_size),
            nn.Flatten(2))
        #CLS token
        self.cls_token = nn.Parameter(torch.randn(1, 1, dim))
        #Positions
        self.pos = PositionalEncoding(dim)
        #Normalización
        self.norm = nn.LayerNorm(dim) #BatchNorm1d(dim)
        self.norm1 = nn.LayerNorm(dim) #BatchNorm1d(dim)
        #Cabezas de atención
        self.att =  nn.ModuleList([deepcopy(SelfAttention(dim)) for _ in range(heads)])
        #Lineal
        self.lin = nn.Linear(heads*dim, dim, bias=False)
        self.ffw = nn.Sequential(nn.Linear(dim, mlp_dim), nn.GELU(),
                                nn.Linear(mlp_dim, num_classes),nn.Softmax(1))
        #Dropouts
        self.drop1 = nn.Dropout(p=p)
        self.drop2 = nn.Dropout(p=p)

    def forward(self, img):
        #Embedding
        e = self.to_patch_embedding(img).transpose(-1,-2)
        cls_tokens = self.cls_token.expand(e.shape[0], -1, -1)
        e = torch.cat((cls_tokens, e), dim=1)
        e += self.pos(e)
        
        #Suma y normalización
        x = self.norm(e) #.transpose(-1,-2))
        x = x + e
        x = self.drop1(x)
        
        #Self-attention
        head_att = [head(x) for head in self.att]
        self.att_weights = [head[1] for head in head_att]
        heads = [head[0] for head in head_att]
        multi_heads = torch.cat(heads, dim=-1)
        h = self.lin(multi_heads)
        
        #Suma y normalización
        h = self.norm1(h) #.transpose(-1,-2))
        h = h + x
        h = self.drop2(h)
        
        #Clasificación
        cls = h[:, 0]
        pred = self.ffw(cls)
        
        return pred
```

## Aplicación del modelo ViT

En primer lugar, definimos el modelo con los hiperparámetros correspondientes. Ya que trabajaremos con el dataset MNIST, indicamos que se usarán patches de tamaño 7, lo que generará un total de 16 pathes en tanto $16 = \frac{28*28}{7*7}$. También definimos el uso de cabezas de atención, las dimensiones y los canales.

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = ViT(image_size=28, patch_size=7, num_classes=10, dim=128, mlp_dim=256, heads=3, channels=1, p=0.3).to(device)
model.load_state_dict(torch.load('weights/model3h3p9.model', weights_only=True)) #, map_location='cpu'))
model
```

Ahora cargamos los datos del MNIST, tanto para entrenamiento, como evaluación. Usamos lotes de un tamaño definido.

```python
batch_size = 32
epochs = 10

# Data transformations (Resize for ViT, ToTensor, Normalize)
transform = transforms.Compose([
    transforms.Resize(28), 
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

train_dataset = datasets.MNIST('./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST('./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

model.eval()
model(train_loader.dataset[2][0].reshape(1,1,28,28))
```

Para el entrenamiento utilizamos el <a href="https://victormijangosdelacruz.github.io/MecanismosAtencion/html/08Noam.html">Optimizador Noam</a> con la entropía cruzada como función objetivo en tanto estamos en una tarea de clasificación. Podemos usar regularización weight decay.

```python
criterion = nn.CrossEntropyLoss()
optimizer = NoamOptimizer(model.parameters(), model.dim, init_lr=0.01, decay=1e-3, warmup=10000)

model.train()
for epoch in range(epochs):
    for data, target in tqdm(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        
    print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')
    torch.save(model.state_dict(), f'weights/model3h3p{epoch+0}.model')
```

## Evaluación

Finalmente, realizamos la evaluación del modelo aprovechando la paquetería de sklearn.

```python
from sklearn.metrics import classification_report

model.eval()
preds = []
y_test = []
for data_t, target_t in test_loader:
    pred = model(data_t).argmax(1)
    preds += pred.tolist()
    y_test += target_t.tolist()

print( classification_report(y_test, preds) )
```

## Visualización de atención

La atención se puede visualizar de forma similar a los transformers. Solo debemos interpretar que las relaciones se dan entre los pathces. De tal forma que los patches tendrán diferente tipo de relación según el peso de atención

```python
from seaborn import heatmap as hm
import matplotlib.pyplot as plt

test, y_test = test_loader.dataset[100]

for att in model.att_weights:
    hm(att.detach().numpy()[0])
    plt.show()
```

Ya que el vector $cls$ es el que codifica la información que se clasifica, las relaciones de los patches con este son significativos en tanto representan las regiones en que el modelo pone atención para realizar la clasificación. Podemos visualizar cómo es que los pesos de atención se ven sobre la imagen.

```python
model(test.reshape(1,1,28,28))

atts = model.att_weights
i = 0
for head in model.att_weights:
    m = head[0][0][1:].reshape(4,4).detach().numpy()
    kernel = m.repeat(7, axis=0).repeat(7, axis=1)

    f, axarr = plt.subplots(1,2)
    plt.title(f'Cabeza {i}')
    axarr[0].imshow(test.numpy().reshape(28,28),cmap='Greys')
    axarr[1].imshow(test.numpy().reshape(28,28)*kernel,cmap='Greys')
    plt.show()
    i += 1
```
